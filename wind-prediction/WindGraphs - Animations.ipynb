```python
from matplotlib.animation import FuncAnimation

from plot import *
```

```python
def animate_interactive_variable_at_time(filename: str,
                                         variable: str,
                                         time: int,
                                         folder: str = "compressed",
                                         output_dir: str = "assets/graphs",
                                         file_format: str = "mp4",
                                         fps: int = 10,
                                         interpolation_levels: int = 8,
                                         **kwargs):

    px = 1 / plt.rcParams["figure.dpi"]
    fig, ax1, ax2 = create_1x2_plot(title="", figsize=(1728 * px, 1080 * px), width_ratios=(98, 2))

    def animate(i):
        print(f"Frame {i}/{frames}")

        ax1.clear()
        ax2.clear()

        i /= interpolation_levels
        if i > 71:
            i = 142 - i

        plot_variable_at_time_and_level(filename, variable, time, i, folder, fig_ax1_ax2=(fig, ax1, ax2), **kwargs)

    frames = 71 * interpolation_levels
    if file_format == "gif":
        frames *= 2

    anim = FuncAnimation(fig, animate, frames=frames, interval=1, blit=True)
    output_file = f"{variable}-{filename.split('.')[-2]}-{1 + time * 3:0>2}:30"
    anim.save(f"{output_dir}/{output_file}-{interpolation_levels}.{file_format}", fps=fps)
```

```python
animate_interactive_variable_at_time("MERRA2_100.tavg3_3d_asm_Nv.19800101.nc4", "U", time=0,
                                     folder="compressed/wind", lat_start=70, lat_end=90)
```
